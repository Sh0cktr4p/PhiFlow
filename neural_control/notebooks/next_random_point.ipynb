```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
knots_x = np.linspace(-1., 5, 7)
randomGenerator = np.random.RandomState()
knots_y = []
amplitude = 5.0
for i, knot_x in enumerate(knots_x): 
    if i == 0: knots_y.append(randomGenerator.uniform(-2., 2.))
    if i > 0: knots_y.append(randomGenerator.uniform(knots_y[0]-0.5*amplitude, knots_y[0]+0.5*amplitude) )
```

```python
plt.plot(knots_x, knots_y)
```
