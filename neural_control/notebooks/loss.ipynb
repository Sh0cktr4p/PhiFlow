```python
import numpy as np 

a = np.arange(31)

def smoothstep(x, nx, step_size): 
    nstep = int(step_size * nx) 
    # Check if x is in the left, middle or right region 
    if x < nstep: xx = 1- x / nstep 
    elif x >= nx-nstep: xx = (x - (nx - nstep ) + 1) / nstep  
    else: xx = 0


    return 3*xx**2 - 2*xx**3

b = []
for x in a:
    b.append(smoothstep(x, len(a), 0.2) )
import matplotlib.pyplot as plt 
plt.plot(a,b)
```
